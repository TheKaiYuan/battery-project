# 容量变化

```python
def GetCapacity(discharge_data):
    data_list = list(discharge_data.keys())
    length = len(data_list)
    capacity = []
    for i in range(length):
        index = data_list[i]
        capacity.append(discharge_data[index]["Capacity"])
    return capacity
```

```python
capacity05 = GetCapacity(discharge_data05)
capacity06 = GetCapacity(discharge_data06)
capacity07 = GetCapacity(discharge_data07)
capacity18 = GetCapacity(discharge_data18)
```

```python
plt.figure(figsize=(15,5), facecolor = "grey")
plt.plot(capacity05, color = "red", label = "05")
plt.plot(capacity06, color = "blue", label = "06")
plt.plot(capacity07, color = "green", label = "07")
plt.plot(capacity18, color = "purple", label = "18")
plt.xlabel("period")
plt.ylabel("capacity")
plt.legend()
plt.show()
```

```python
begin_capacity = [capacity05[0][0], capacity06[0][0], capacity07[0][0], capacity18[0][0]]
```

```python
plt.figure(figsize=(10,5), facecolor = "grey")
plt.bar(range(len(begin_capacity)), begin_capacity, tick_label = tick_label, color = ["red", "blue","green","purple"])
plt.title("begin_capacity")
plt.show()
```

```python
def trans2feature(data):
    future_name = ["period", "avg_voltage","terminate_voltage", "min_temp", "max_temp", "avg_temp", 
               "avg_voltage_load", "terminate_voltage_load", "begin_capacity", "capacity"]
    out = []
    period = 0
    keys = list(data.keys())
    begin_capacity = data[keys[0]]["Capacity"][0]
    for key in keys:
        one_period = data[key]
        avg_voltage = np.average(one_period["voltage_battery"])
        terminate_voltage = min(one_period["voltage_battery"])
        min_temp = min(one_period["temp_battery"])
        max_temp = max(one_period["temp_battery"])
        avg_temp = np.average(one_period["temp_battery"])
        avg_voltage_load = np.average(one_period["voltage_load"])
        i = len(one_period["voltage_load"]) - 1
        while one_period["voltage_load"][i] < 0.1:
            i = i - 1
        terminate_voltage_load = one_period["voltage_load"][i]
        capacity = one_period["Capacity"][0]
        one_data = [period, avg_voltage, terminate_voltage, min_temp, max_temp, avg_temp, avg_voltage_load, terminate_voltage_load, 
                    begin_capacity, capacity]
        out.append(one_data)
        period = period + 1
    return pd.DataFrame(out, columns = future_name)
```

```python
D05 = trans2feature(discharge_data05)
D06 = trans2feature(discharge_data06)
D07 = trans2feature(discharge_data07)
D18 = trans2feature(discharge_data18)
```

```python
features = pd.concat([D05, D06, D07, D18])
```

```python
features = features.reset_index(drop = True)
```

```python
features.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.model_selection import train_test_split
from sklearn import metrics
```

```python
label = features["capacity"]
fea = features.drop(labels = ["capacity"], axis = 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(fea, label, test_size = 0.2, random_state = 5)
```

# 线性回归

```python
LR = LinearRegression()
LR.fit(x_train, y_train)
```

```python
y_pred = LR.predict(x_test)
```

```python
print("MSE:",metrics.mean_squared_error(y_test, y_pred))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

# SVR

```python
svr = SVR()
svr.fit(x_train, y_train)
```

```python
y_pred2 = svr.predict(x_test)
```

```python
print("MSE:",metrics.mean_squared_error(y_test, y_pred2))
print("RMSE:",np.sqrt(metrics.mean_squared_error(y_test, y_pred2)))
```
